Load data for fitting

```python
import pandas as pd

maxQ = pd.read_csv("drive/MyDrive/Colab Notebooks/CE6280/Data/Table6.2.csv")
```

Fit LN2 and LN3 distributions using MOM, MLE, and Lmom. Code for this is in utils.py.

```python
methods = ["MOM", "MLE", "Lmom"]
npars = [2, 3]

for method in methods:
  for npar in npars:
    distfit = LogNormal()
    distfit.fit(maxQ["Discharge"],method, npar)
    q100 = distfit.findReturnPd(100)
    distfit.plotHistPDF(maxQ["Discharge"], 0, 4500, "LN" + str(npar) + " " + str(method) + " Fit")
    print("LN%d %s mu: %0.2f" % (npar, method, distfit.mu))
    print("LN%d %s sigma: %0.2f" % (npar, method, distfit.sigma))
    print("LN%d %s tau: %0.2f" % (npar, method, distfit.tau))
    print("LN%d %s 100-yr flood: %0.0f cms" % (npar, method, q100))
    print("\n")
```
